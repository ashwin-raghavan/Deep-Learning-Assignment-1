```python
def calculate_stochastic_gradint_descent_parameters(X_train, y_train, weights, bias):
    # Define parameters
    # Choose values for α, max_iterations, threshold, N
    # Set w and b to a set of valid initial values (w is same size as x(i))
    # Set stopping = False; J_running = 0; J_running_prev = 0; iteration = 0
    
    alpha = 0.01
    max_iterations = 5000
    threshold = 0.001
    N = len(X_train)
    stopping = False
    J_running = 0
    J_running_prev = 0
    iteration = 0
    while not stopping:
        # Select a single example from training set at random    
        idx = np.random.randint(0, N)
        x_i = X_train.iloc[idx]
        y_i = y_train.iloc[idx]    
        # Forward propagation stage
        # Calculate z   
        z = np.dot(x_i, weights) + bias # source: https://stackoverflow.com/questions/5919530/what-is-the-pythonic-way-to-calculate-dot-product
    
        # Calculate y_hat by applying sigmoid
        y_hat = 1 / (1 + np.exp(-z))
    
        # Calculate loss or J_current
        # from the lectures loss = - (y(i) * log(y_hat(i)) + (1 - y(i)) * log(1 - y_hat(i)))   
        y_hat = np.clip(y_hat, 1e-15, 1 - 1e-15)
        J_current = -(y_i * np.log(y_hat) + (1-y_i) * (np.log(1-y_hat)))
    
        
        
    
        #Calculate partial derivates. 
        # From the lecture we have to get the error that is, y_hat(i) - y(i)
        # After this we will calculate the partial derivatives as :
        # Delta w_j = (y_hat(i) - y(i)) * x_j(i)
        # Delta b   = (y_hat(i) - y(i))
        error = y_hat - y_i
        delta_w = error * x_i # gives us vector of delta weight one for each feature
        delta_b = error # Scalar value
        
        # Loop again over the j elements of w:
        # w_j -= alpha * delta_w
        # b   -= alpha * delta_b
        weights -= alpha * delta_w
        bias -= alpha * delta_b
    
        # Check stopping criteria
        iteration += 1
        J_running += J_current
      
        if iteration > max_iterations:
            stopping = True # Failed to converge
            print(f"Stopped: Reached max iterations ({iteration})")
            
        if (iteration % N) == 0: # Have done N iterations: test for convergence on the batch
            # Compare J_running with J_running_prev: if less than threshold, stopping = True    
            if (J_running - J_running_prev) < threshold:          
                stopping = True
                print(f"Stopped: Converged at iteration {iteration}. Cost diff: {J_running - J_running_prev}")
    
            # Set J_running_prev = J_running; J_running = 0
            J_running_prev = J_running
            J_running = 0

    return weights, bias
    
    




    
    
```

```python
weights, bias = calculate_stochastic_gradint_descent_parameters(X_train, y_train, weights, bias)
print(f"Final Weights: {weights}")
print(f"Final Bias: {bias}")
    
```

```python
def test_stochastic_gradint_descent_parameters(X_test, y_test, weights, bias):   
    X_test['z(x)'] = X_test[features].dot(weights) + bias     
    
    # Apply Sigmoid to z(x)
    def sigmoid(z):
        return 1 / (1 + np.exp(-z))

    X_test['y_hat'] = X_test['z(x)'].apply(sigmoid) 

    def get_y_pred(y_hat):
        if y_hat >= 0.5: return 1

        return 0
    

    X_test['y_pred'] = X_test['y_hat'].apply(get_y_pred) 

    accuracy = np.mean( X_test['y_pred'] == y_test) # count the number of 1's. if everythign is correct then number 1s is equal to the number rows. mean of that is 100%
    print(f"Test Accuracy: {accuracy * 100:.2f}%")

    

```
